**Map**

`This function is used to access the values one after other (if we use map there is not need to use loops )`

`syntax : map(function_name , iterables )`

```python
def sharuk():
  for i in range(11,21):
    if i % 2 == 0:
      print(f"even : {i}")
    else:
      print(f"Odd : {i}")



sharuk() # calling the function
```

**Map with Filter**

```python
sol = map(lambda x : x + 20 , filter(lambda x : x % 2==0 , [10,11,12,14,15,16]))
print(list(sol))
```

**Filter with MAP**

```python
sol = filter(lambda x : x % 2 == 0 , map(lambda x : x + 20 , [10,11,12,14,15,16]))
print(list(sol))
```

**Reduce with Map with Filter**

```python
sol = functools.reduce(lambda x,y : x+ y , map(lambda x : x + 20 , filter(lambda x : x%2 == 0 , [10,11,12,14,15,16])))
print(sol)
```

**Comprehensions**

`It is used for reducing the lines of code `

we have 2 types of comprehensions


 - `List comprehensions `
 - `Dictionary Comprehensions `

**List Comprehensions**

- `syntax 1: one forloop and one print condition `

       [<print_area> forloop]


- `syntax 2:  one forloop and one if condition and if print `

      [<print_area> forloop if_condition]
    
- `syntax 3: one forloop and one if condition and if print and else condition and else print `

      [<if_print> <if_condition> <else_condition> <else_print> forloop]

```python
# step : 1

for i in range(1,6):
  print('Good Morning')
```

```python
print(['Good Morning' for i in range(1,6)])
```

```python
# step:2

for i in range(11,16):
  if i % 2 == 0:
    print(f'Even : {i}')
```

```python
print([f'Even : {i}' for i in range(11,16) if i % 2 == 0])
```

```python
# step:3

for i in range(11,16):
  if i % 2 == 0:
    print(f'Even : {i}')
  else:
    print(f'Odd : {i}')
```

```python
print([f'Even : {i}' if i % 2 == 0  else  f'Odd : {i}' for i in range(11,16)])
```

**Dictionary Comprehensions**

```python
# step:1

fruits = {'apple':100 , 'banana':200 , 'carrot':300}

for i in fruits:
  fruits[i] = fruits[i] * 2

print(fruits)
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

print({i:fruits[i]*2 for i in fruits})
```

```python
# step:2

fruits = {'apple':100 , 'banana':200 , 'carrot':300}

d = {}
for i in fruits:
  if fruits[i] > 150:
    d[i] = fruits[i]
print(d)
```

```python
print({i:fruits[i] for i in fruits if fruits[i] > 150})
```

```python
# step:3

fruits = {'apple':100 , 'banana':200 , 'carrot':300}


for i in fruits:
  if fruits[i] > 150:
    fruits[i] = fruits[i] * 3
  else:
    fruits[i] = fruits[i] * 2


print(fruits)
```
