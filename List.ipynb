**Manually Increasing the Memory size and decreasing Not possible in List concept (so please take help of List concept Built_in_functions)**

**List Built_in_functions**


**Increase**
- `append `
- `extend `
- `insert `

**Decreasing**
- `pop `
- `remove `
- `clear `

**Some Other Operations**
- `copy `
- `count `
- `index `
- `sort `
- `reverse `


```python
# append = this function will take only 1 input from user and add to the end of
# the existing List

f = [10,15,2,9,84,36] # save in the Memory
print(f'Before Append : {f}')


f.append(500) # value will add (auto update)

print(f'After Append : {f}') # caling from Memory
```

```python
# extend = this function will take multiple inputs from user and add to the end of
# the existing List

f = [10,15,2,9,84,36] # save in the Memory
print(f'Before Extend : {f}')


f.extend([500,1000,1500])     # value will add (auto update)

print(f'After Extend : {f}') # caling from Memory
```

```python
# Insert = will take value from user and add where you want in the existing list


f = [10,15,2,9,84,36] # save in the Memory
print(f'Before Insert : {f}')

f.insert(3 , 'america')
f.insert(5 , 'usa')

print(f'After Insert : {f}')
```

```python
# pop:
# Rule:1 = this function defaultly will remove last value from the existing List
# Rule:2 = and if you pass any input that index value will be removed

f = [10,15,2,9,84,36] # save in the Memory
print(f'Before Pop : {f}')


f.pop() # last value removed (auto updated)


print(f'After Pop : {f}')
```

```python
# pop:
# Rule:1 = this function defaultly will remove last value from the existing List
# Rule:2 = and if you pass any input that index value will be removed

f = [10,15,2,9,84,36] # save in the Memory
print(f'Before Pop : {f}')


f.pop(3) # 3rd index value removed (auto updated)


print(f'After Pop : {f}')
```

```python
# remove = pass value it wil remove


f = [10,15,2,9,84,36] # save in the Memory
print(f'Before remove : {f}')


f.remove(84) # (auto updated)


print(f'After remove : {f}')
```

```python
a = [10,20,30] # save in the Memory

del a

print(a) # calling from Memory
```

```python
# clear = will remove all vaues from varibale
# but varibale is alive in the Memory


f = [10,15,2,9,84,36] # save in the Memory
print(f'Before clear : {f}')


f.clear() # auto update


print(f'After clear : {f}')
```

**From Above Operations we can say that Python List is Mutable**

```python
# count

a = [10,20,10,30,40]

print(a.count(40))
```

```python
# index

a = [10,20,10,30,40]

print(a.index(10 , 1))
```

```python
# reverse [::1]

f = [10,15,2,9,84,36] # save in the Memory
print(f'Before reverse : {f}')


f.reverse() # [::-1] (auto update)


print(f'After reverse : {f}')
```

```python
# sort = ascending and descending

f = [10,15,2,9,84,36] # save in the Memory
print(f'Before sort : {f}')

f.sort() # ascending order (auto update)

print(f'After sort : {f}')
```

```python
# sort = ascending and descending

f = [10,15,2,9,84,36] # save in the Memory
print(f'Before sort : {f}')

f.sort(reverse=True) # descending order (auto update)

print(f'After sort : {f}')
```

```python
# copy

a = [10,20,30]
b = a # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(100)

print(a)
print(b)
```

```python
# copy

a = [10,20,30]
b = a.copy() # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(100)

print(a)
print(b)
```
