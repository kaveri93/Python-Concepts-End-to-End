```python
import sys
class PRIMENUMBER:
  # n = 7
  def __init__(self,n):
    try:
      self.n = n
    except Exception as e:
      error_ty,er_msg,er_line = sys.exec_info()
      print(f"Error in Line no : {er_line} due to : {er_msg}")

  def logic_(self):
    try:
      c = 0
      for i in range(1,self.n + 1):
        if self.n % i == 0:
          c = c + 1
      if c == 2:
        print(f"{self.n} : is Prime Number")
      else:
        print(f"{self.n} : is Not Prime Number")
    except Exception as e:
      error_ty,er_msg,er_line = sys.exec_info()
      print(f"Error in Line no : {er_line} due to : {er_msg}")

obj = PRIMENUMBER(11)
obj.logic_()
```

**Inheritance**

`Means it is a Parent and child relation `

There are 5 types of Inheritance

- `Single Inheritance `
- `Multiple Inheritance `
- `Multilevel Inheritance `
- `Hierarichal Inhertiance `
- `Hybrid Inheritance `

```python
# single Inheritance

class PARENT:
  def building(self):
    print("I have My own Building")


class CHILD(PARENT):
  def job(self):
    print('I have my own Job')


anil = PARENT()
anil.building()
```

```python
# single Inheritance

class PARENT:
  def building(self):
    print("I have My own Building")


class CHILD(PARENT):
  def job(self):
    print('I have my own Job')

kiran = CHILD()
kiran.job()
kiran.building()
```

```python
def fun(a = None , b = None):
  if a == None and b == None:
    print("I am from USA")
  elif a == None and b != None:
    print("I am from UK")
  elif a != None and b == None:
    print("I am from Paris")
  else:
    print("I am from India")


fun(10,20)
```

```python
# Abstraction
from abc import ABC,abstractmethod

class stage_1(ABC):
  def bus(self):
    print("Logics for Bus")

  def railway(self):
    print("Logics for railway")

  @abstractmethod
  def banking_connections(self):
    pass

class Stage_2(stage_1):
  def banking_connections(self):
    print("New Logics for Baking")


obj = Stage_2()
obj.banking_connections()
```
