```python
# array -> will create the data in the form of dimesion (1D | 2D | 3D)

a = np.array( [10 , 20 , 30,40] )

print(a) # calling from the Memory

print(a.ndim) # ndim (number of dimesions)

print(a.shape)  # shape (rows and columns)
```

```python
# arange will work like range in forloop only

a = np.arange(11,17) # it will work with (n-1) and the output will be in the form on 1D
print(a)
print(a.ndim)
```

```python
print(a[3])
```

```python
# 1D to 2D

a = np.arange(11,17)
print(a)
print(a.ndim)
print(a.shape)

print('=========================')

b = a.reshape(2,3)
print(b)
print(b.ndim)
print(b.shape)
```

```python
a = np.arange(11,17).reshape(2,3)
print(a)
print(a.ndim)
print(a.shape)
```

```python
# reshape(-1,1)

a = np.array([10,11,12,13])
print(a)
print(a.ndim)
print(a.shape)

print("==========================")

b = a.reshape(-1,1)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# reshape(1,-1)

a = np.array([10,11,12,13])
print(a)
print(a.ndim)
print(a.shape)

print("==========================")

b = a.reshape(1,-1)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# numpy -> random -> randint
# give range it will pick random integer values

a = np.random.randint(10,50 , 5)

a
```

```python
# numpy -> random -> rand
# random -> function will gives values between 0 to 1

a = np.random.random((4,4))
a
```

```python
# numpy -> random -> randn
# randn -> function will gives values between -3 to 3

a = np.random.randn(4,4)
a
```

```python
# ravel -> any dimension data into 1D

a = np.array([  [10,9,8] , [7,4,11] , [13,2,6]   ])
print(a)
print(a.ndim)
print(a.shape)
print(a.size)

print('===================')

b = np.ravel(a)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# transpose
a = np.array([  [10,9,8] , [7,4,11] , [13,2,6]   ])
print(a)

print("===================")

b = np.transpose(a)
print(b)
```

```python
# concatinate
a = np.array([  [10,9,8] , [7,4,11] , [13,2,6]   ])
print(a)
print('=============================')
b = np.array([ [1,2,3] , [4,5,6] , [10,9,8]  ])
print(b)

print('================================')

c = np.concatenate((a,b))
print(c)
```
