**File Handling**

- `This concept is used to deal with files (.txt | .xlsx | .csv etc....)`
- `Using this concept we can `
    
           - `write the data into a file -> w`
           - `add the data into a file -> a`
           - `read the data from the file -> r`

**Mode w**

- `step : 1 -> If the File is not there Mode w will create the file and will write the data into the file `

- `step : 2 -> If the File is Available but File is Empty Mode w will open and will write the data into the file `

- `step : 3 -> If the File is Available and data is also Available mode w will override the data with new data `

```python
a = open('sharuk.txt' , 'w')

a.write("Data science with GEN AI")

a.close()
```

```python
a = open('kiran.txt' , 'w')

a.write("Data science with GEN AI")

a.close()
```

```python
a = open('sharuk.txt' , 'w')

a.write("USA")

a.close()
```

**Mode a**

- `step : 1 -> If the File is not there Mode a will create the file and will write the data into the file `

- `step : 2 -> If the File is Available but File is Empty Mode a will open and will write the data into the file `

- `step : 3 -> If the File is Available and data is also Available mode a will add the data`

```python
a = open('sharuk.txt' , 'a')

a.write("Data science with GEN AI")

a.close()
```

```python
a = open('kiran.txt' , 'a')

a.write("Data science with GEN AI")

a.close()
```

```python
a = open('sharuk.txt' , 'a')

a.write(" USA")

a.close()
```

**Mode r**

```python
a = open('kiran.txt' , 'r')

print(a.read())

a.close()
```

```python
with open('kiran.txt' , 'r') as a:
  print(a.read())
```

```python
b = open("states.txt" , "w")

b.write("TG | AP | KN ")
```

```python
with open("states.txt" , "w") as b:
  b.write("TG | AP | KN ")
```

json (java script object notation) -> is used to save the dictionary (key_value ) pairs

```python
import json
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300} # dict
# dump is used to sent the dict into a json file
with open("project_1_info.json","w") as f:
  json.dump(fruits , f)
```

```python
# load is used to take dict from the json file


with open("project_1_info.json" , "r") as f:
  result = json.load(f)

print(result)
```

```python
# dumps = convert the dict into str

fruits = {'apple':100 , 'banana':200 , 'carrot':300} # dict

r = json.dumps(fruits) # convert dict into str

print(r) # in this varibale we have str (for dict)
print(type(r))
```

```python
# loads = take str (for dict) back to dict
t = json.loads(r)
t
```

**Recurssion**

```python
def fun():
  print("Good Morning")
  fun() #calling


fun()  # calling
```

```python
def fun(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * fun(n - 1)


fun(5) # calling
```
