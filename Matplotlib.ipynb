```python
# scatter plot

india = [1,5,8,3,6,9]
usa = [4,1,5,2,3,6]


plt.scatter(x = india , y = usa) # only math operations

plt.show()
```

```python
# scatter plot

india = [1,5,8,3,6,9]
usa = [4,1,5,2,3,6]

plt.figure(figsize=(5,3))  # (w , h)
plt.title("Scatter Plot Visualization")
plt.xlabel("India_values")
plt.ylabel("Usa_values")

plt.scatter(x=india,y=usa,marker='*',color='green')

plt.show()
```

```python
# scatter plot

india = [1,5,8,3,6,9]
usa = [4,1,5,2,3,6]
uk = [5,3,6,8,7,2]

plt.figure(figsize=(5,3))  # (w , h)
plt.title("Scatter Plot Visualization")
plt.xlabel("India_values")
plt.ylabel("Usa_and_uk_values")

plt.scatter(x=india,y=usa,marker='*',color='green')
plt.scatter(x=india,y=uk,marker='s',color='black')


plt.show()
```

```python
# line chart

india = [1,5,8,3,6,9]
usa = [4,1,5,2,3,6]
uk = [5,3,6,8,7,2]

plt.figure(figsize=(5,3))  # (w , h)
plt.title("line chart Visualization")
plt.xlabel("India_values")
plt.ylabel("Usa_and_uk_values")

plt.plot(india,usa,label='india-usa')
plt.plot(india,uk,label='india-uk')

plt.legend()

plt.show()
```

```python
# subplots

india = [1,5,8,3,6,9]
usa = [4,1,5,2,3,6]
uk = [5,3,6,8,7,2]

plt.figure(figsize=(8,3))  # (w , h)


plt.subplot(1,2,1)
plt.title('india-usa')
plt.xlabel('india')
plt.ylabel('usa')
plt.plot(india,usa,color='red',marker='*')


plt.subplot(1,2,2)
plt.title('india-uk')
plt.xlabel('india')
plt.ylabel('uk')
plt.plot(india,uk,color='black',marker='d')

plt.show()
```

```python
# subplots

india = [1,5,8,3,6,9]
usa = [4,1,5,2,3,6]
uk = [5,3,6,8,7,2]

plt.figure(figsize=(5,5))  # (w , h)


plt.subplot(2,1,1)
plt.title('india-usa')
plt.xlabel('india')
plt.ylabel('usa')
plt.plot(india,usa,color='red',marker='*')


plt.subplot(2,1,2)
plt.title('india-uk')
plt.xlabel('india')
plt.ylabel('uk')
plt.plot(india,uk,color='black',marker='d')

plt.show()
```

```python
# Histogram = is used to find the number of data points between the intervals

plt.figure(figsize=(5,3))  # (w , h)
age = [1,4,7,9,3,11,17,21,27,23,26,31,33,39]

plt.hist(x = age , bins = 4 , color = 'r')

plt.show()
```

```python
# Bar chart

company = ['Google','Meta','IBM','Infosys']
revenue = [1200 , 1600 , 900 , 600]

plt.figure(figsize=(5,3))  # (w , h)
plt.bar(company , revenue , color = ['r','y','g','black'])

plt.show()
```

```python
# Bar chart

company = ['Google','Meta','IBM','Infosys']
revenue = [1200 , 1600 , 900 , 600]

plt.figure(figsize=(5,3))  # (w , h)
plt.barh(company , revenue , color = ['r','y','g','black'])

plt.show()
```

```python
# Pie chart

company = ['Google','Meta','IBM','Infosys']
revenue = [1200 , 1600 , 900 , 600]

plt.figure(figsize=(5,3))  # (w , h)


plt.pie(revenue , labels = company)

plt.show()
```

```python
# Pie chart

company = ['Google','Meta','IBM','Infosys']
revenue = [1200 , 1600 , 900 , 600]

plt.figure(figsize=(5,3))  # (w , h)


plt.pie(revenue , labels = company , shadow=True)

plt.show()
```

```python
# Pie chart

company = ['Google','Meta','IBM','Infosys']
revenue = [1200 , 1600 , 900 , 600]

plt.figure(figsize=(5,3))  # (w , h)


plt.pie(revenue , labels = company , shadow=True , autopct='%.2f')

plt.show()
```
