**Built_in_functions**

- `get `
- `update `
- `pop `
- `popitem`
- `copy `
- `clear `
- `keys `
- `values `
- `items `

```python
# get

fruits = {'apple':100 , 'Banana':200 , 'Carrot' : 300}

print(fruits['Banana'])
```

```python
# get

fruits = {'apple':100 , 'Banana':200 , 'Carrot' : 300}

print(fruits.get('Banana'))
```

```python
# update

fruits = {'apple':100 , 'Banana':200 , 'carrot':300}
print(f'Before : {fruits}')

fruits['grapes'] = 500

print(f'After : {fruits}')
```

```python
# update

fruits = {'apple':100 , 'Banana':200 , 'carrot':300}
print(f'Before : {fruits}')

fruits.update({'grapes':500})

print(f'After : {fruits}')
```

```python
# pop = pass column it will remove

fruits = {'apple':100 , 'Banana':200 , 'carrot':300}
print(f'Before : {fruits}')

del fruits['Banana']

print(f'After : {fruits}')
```

```python
# pop = pass column it will remove

fruits = {'apple':100 , 'Banana':200 , 'carrot':300}
print(f'Before : {fruits}')

fruits.pop('Banana')

print(f'After : {fruits}')
```

```python
# popitem = remove last column

fruits = {'apple':100 , 'Banana':200 , 'carrot':300}
print(f'Before : {fruits}')

fruits.popitem()

print(f'After : {fruits}')
```

```python
# clear

fruits = {'apple':100 , 'Banana':200 , 'carrot':300}
print(f'Before : {fruits}')

fruits.clear()

print(f'After : {fruits}')
```

```python
fruits = {'apple':100 , 'Banana':200 , 'carrot':300}
b = fruits

print(fruits)
print(b)

print(id(fruits))
print(id(b))

fruits['apple'] = 500

print(fruits)
print(b)
```

```python
fruits = {'apple':100 , 'Banana':200 , 'carrot':300}
b = fruits.copy()

print(fruits)
print(b)

print(id(fruits))
print(id(b))

fruits['apple'] = 500

print(fruits)
print(b)
```
