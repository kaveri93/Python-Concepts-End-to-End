**Python tuple concept is Mutable or Immutable -> Prove it**

```python
# try to increase the Memory size

a = (10,50,2) # save in the Memory

a[3] = 500

print(a) # calling from the Memory
```

```python
# try to decrease the Memory size

a = (10,50,2) # save in the Memory

del a[1]

print(a) # calling from the Memory
```

`From above Operations we can say that Python tuple is Immutable`
