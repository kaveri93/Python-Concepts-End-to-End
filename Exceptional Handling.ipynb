**Exceptional Handling**

`Exceptional Handling concept is used to optimize the error but not to solve the error `


**Some Common Error's**

- `Syntax Error `
- `ZeroDivisionError`
- `Name Error `
- `Value Error `
- `Indentation Error `
- `FileNotFound Error `

```python
# syntax error

10a = 50 # save in the Memory
print(10a) # calling from the Memory
```

```python
# ZeroDivisionError

5 / 0
```

```python
# NameError

city = 'hyd'
print(CITY)
```

```python
# valueError

a = 'kamal'
b = int(a)
print(b)
```

```python
# indentation error

for i in range(1,11):
print(i)
```

```python
# FIleNotFound Error

with open('sharuk.txt','r') as f:
  print(f.read())
```

- `From the Above Error's if we see we are getting around 8 to 10 lines of error msg `
- `But there is no need of 8 to 10 lines of error msg but defaulty python gives 8 to 10 lines `
- `Using Exceptional Handling we can reduce the 8 to 10 lines of error into a single line (optimizing the error )`
- `Exceptional Handling works with try block and except block `

```python
try:
  a = 10
  b = 20
  print(a / b)


except Exception as e:
  print(e)
```

```python
import sys

try:
  a = 10
  b = 20
  print(a / b)


except Exception as e:
  print(f'Error Msg By try Block : {e}')
  error_type,error_msg,error_line = sys.exc_info()
  print(f'Error Type was : {error_type}')
  print(f'Error Msg was : {error_msg}')
  print(f"Error is coming from Line no : {error_line.tb_lineno}")
```
