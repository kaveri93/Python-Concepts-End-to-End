**Difference Between Local varibales and Global Varibales**

```python
r = 100 # global varibale

def fun():
  print(f"I am from Inside the function : {r}")

fun() # calling
print(f"I am from outside the function : {r}")
```

```python
def fun():
  pr = 100 # local variable
  print(f"I am from Inside the function : {pr}")

fun() # calling
print(f"I am from outside the function : {pr}")
```

```python
def fun():
  global pr
  pr = 100 # local variable to global varibale
  print(f"I am from Inside the function : {pr}")

fun() # calling
print(f"I am from outside the function : {pr}")
```

**Small Project on Functions**

```python
def addition(g,h):
  return g + h


def subtraction(g,h):
  return g - h


def multiplication(g,h):
  return g * h


def division(g,h):
  return g / h


def common(var,a,b):
  if var == 'add':
    sol = addition(a,b)
    return sol
  elif var == 'sub':
    sol = subtraction(a,b)
    return sol
  elif var == 'mul':
    sol = multiplication(a,b)
    return sol
  elif var == 'div':
    sol = division(a,b)
    return sol
  else:
    return f'Issue in Input'



a = int(input())  # 5
b = int(input())  # 7
c = input()       # mul

f = ['add','sub','mul','div']
for i in f:
  if i == c:
    final_ans = common(i,a,b)

print(final_ans)
```

**Lambda Functions**

```python
# Lambda Function is known as Ananymous Function
# Ananymous Function (A function without a Function Name)

# syntax of lambda function = lambda arguments : Expressions
```

```python
a = 10
b = 20
print(a + b)
```

```python
def fun(a,b):
  print(a + b)


fun(10,20) # calling
```

```python
c = lambda a,b : a + b


print(c(10,20))
```

```python
c = lambda a , b : a * b


print(c (5,7))
```

```python
n = 2

for i in range(1,11):
  print(f'{n} * {i} = {n * i}')
```

```python
def fun(n):
  for i in range(1,11):
    print(f'{n} * {i} = {n * i}')



fun(2)
```

```python
c = lambda a , b : a * b


n = 2
for i in range(1,11):
  print(f'{n} * {i} = {c(n,i)}')
```
