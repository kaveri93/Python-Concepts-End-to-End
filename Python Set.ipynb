**Rules of Python Set concept**

- `Python Set concept is going to represent with -> {}`
- `Python Set concept is an unordered Pair `

- `In Python set concept there will be no index if there is no index there will be no slicing and no skipping concept `

- `Python Set concept will not allow duplicates `

- `Python Set concept is Mutable But it will not allow mutable data types inside it `

**Built_in_functions**

**Adding**
- `add`
- `update `

**Decreasing**
- `remove `
- `discard`
- `pop `

**Casual Usage**
- `clear `
- `copy `
- `union `
- `intersection `

```python
# add = this function will take one input from the user and add randomly somewhere
# in the existing set

a = {10,20,30,50,90} # save in the Memory
print(f'Before add function : {a}')

a.add(100) # auto update

print(f'After add function : {a}')
```

```python
# update = this function will take multiple inputs from the user and add randomly somewhere
# in the existing set

a = {10,20,30,50,90} # save in the Memory
print(f'Before update function : {a}')

a.update([100,500,900]) # auto update

print(f'After update function : {a}')
```

```python
# pop = this function removes one value randomly from the existing set

a = {10,20,30,50,90} # save in the Memory
print(f'Before pop function : {a}')

a.pop() # auto update

print(f'After pop function : {a}')
```

```python
# discard = this function will take one input from the user
# and if value is availabe in existing set it will remove
# or else it will be silent

a = {10,50,90,100}  # save in the Memory
print(f'Before Discard : {a}')

a.discard(90)

print(f'After Discard : {a}')
```

```python
# remove = this function will take one input from the user
# and if value is availabe in existing set it will remove
# or else it will through error

a = {10,50,90,100}  # save in the Memory
print(f'Before Discard : {a}')

a.remove(90)

print(f'After Discard : {a}')
```

`From Above Operations Python Set is Mutable`

```python
# copy

a = {10,20,50}
b = a # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(100)

print(a)
print(b)
```

```python
# copy

a = {10,20,50}
b = a.copy() # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(100)

print(a)
print(b)
```

```python
# union
a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.union(b))
```

```python
# intersection
a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.intersection(b))
```
