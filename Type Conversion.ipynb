`Type Conversion Means we are going to convert one data type into other data type `

- `int -> float and int -> str `
- `float -> int and float -> str `
- `str -> int and str -> float `

```python
# int -> float

a = 12 # save in the Memory
print(a) # calling from the Memory
print(type(a)) # checking the data type of a varibale

b = float(a) # actual conversion is going on here
print(b) # calling from the Memory
print(type(b))# checking the data type of a varibale
```

```python
# int -> str

a = 12 # save in the Memory
print(a) # calling from the Memory
print(type(a)) # checking the data type of a varibale

b = str(a) # actual conversion is going on here
print(b) # calling from the Memory
print(type(b))# checking the data type of a varibale
```

**Rule for Str**

- `If we want to convert str into int or str into float we need to maintain number only in between the quotes `

`Ex1: 'india' or 'india123' -> convert into int or float -> not possible `

`Ex2: '12' or '12.33' -> convert into int or float -> Possible `

`we are going to create the relationship between str -> str and str -> int and str -> float inside the print condition`

- `Rule:1 -> Using comma (,) we can create relationship with str to any data type where comma (,) will give one tab space for sure `

- `Rule:2 -> using (+) we can create  relationship with str to any data type where + should have both str only and + will not give one tab space `

- `Rule:3 -> Using Identifiers -> %d -> int %f -> float -> %s -> str `

- `Rule:4 -> using f and {}`

```python
a = 27 # save in the Memory

print('My Name is John and My age is',a)
```

```python
course = 'data analysis'
time = 9.30
duration = 6

# expected output = I joined data analysis course at vihara tech for 9.30 class about 6 months

print('I joined',course,'course at vihara tech for',time,'class about',duration,'months')
```

```python
# rule 2 (+)

a = 27 # save in the Memory

print('My Name is John and My age is '+str(a))
```

```python
course = 'data analysis'
time = 9.30
duration = 6

# expected output = I joined data analysis course at vihara tech for 9.30 class about 6 months

print('I joined '+course+' course at vihara tech for '+str(time)+' class about '+str(duration)+' months')
```

```python
# rule 3 (identifiers)

a = 27 # save in the Memory

print('My Name is John and My age is %d' %(a))
```

```python
course = 'data analysis'
time = 9.30
duration = 6

# expected output = I joined data analysis course at vihara tech for 9.30 class about 6 months

print('I joined %s course at vihara tech for %f class about %d months' %(course,time,duration))
```

```python
course = 'data analysis'
time = 9.30
duration = 6

# expected output = I joined data analysis course at vihara tech for 9.30 class about 6 months

print('I joined %s course at vihara tech for %f class about %d months' %(duration,time,course))
```

```python
# rule 4

a = 27 # save in the Memory

print(f'My name is john and my age is {a}')
```

```python
course = 'data analysis'
time = 9.30
duration = 6

# expected output = I joined data analysis course at vihara tech for 9.30 class about 6 months

print(f'I joined {course} at vihara tech for {time} class about {duration} months')
```
