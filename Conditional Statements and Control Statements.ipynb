```python
# Perfect Number

n = 6
c = []

for i in range(1,n):
  if n % i == 0:
    c.append(i)

if sum(c) == n:
  print(f'{n} : is a Perfect Number')
else:
  print(f'{n} : is Not a Perfect Number')
```

```python
# Perfect Number

n = 6
c = 0 # 6

for i in range(1,n):
  if n % i == 0:
    c = c + i

if c == n:
  print(f'{n} : is a Perfect Number')
else:
  print(f'{n} : is Not a Perfect Number')
```

```python
# Palindrome

a = 121

if a == int(str(a)[::-1]):
  print(f'{a} : is a Palindrome ')
else:
  print(f'{a} : is not a Palindrome ')
```

```python
n = 121
t = n
c = 0

while n > 0:
  res = n % 10
  c = (c * 10) + res
  n = n // 10

if t == c:
  print(f'{t} : is a Palindrome ')
else:
  print(f'{t} : is not a Palindrome ')
```

**Amstrong**

```python
n = 153
t = n
c = 0

while n > 0:
  res = n % 10
  c = c + (res * res * res)
  n = n // 10

if t == c:
  print(f'{t} : is a Amstrong ')
else:
  print(f'{t} : is not a Amstrong ')
```

```python
# strong number

n = 145
f = n
c = 0 # 145

      # 0 > 0
while n > 0:
  res = n % 10  # 1

  t = 1  # 1
  for i in range(1,res+1):
    t = t * i

  c = c + t
  n = n // 10 # 0


if f == c:
  print(f'{f} : is a Strong Number')
else:
  print(f'{f} : is not a strong number')
```

```python
n = 145
f = n
t = 0 # 145

      # 0 > 0
while n > 0:
  res = n % 10 # 1
  s = math.factorial(res) # 1! (1)
  t = t + s
  n = n // 10 # 0


if f == t:
  print(f'{f} : is a Strong Number')
else:
  print(f'{f} : is not a strong number')
```

```python
# Fibonacci Series

n = 8

a = 0
b = 1
print(a)
print(b)

for i in range(n-2):
  c = a + b
  print(c)
  a = b
  b = c
```
