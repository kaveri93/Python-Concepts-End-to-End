**Rules to create a varibale**

- `Rule No:1 -> Varibale Should not start with a Number if you want to start please use Underscore `
- `Rule No:2 -> Varibale should be Single charater or single word Multiple words are not Allowed if you want please use Underscore `
- `Rule No:3 -> Python varibales are case sensitive (capital means capital small means small)`

- `Rule No:4 -> Dont use below punctuations as varibale -> !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~`
