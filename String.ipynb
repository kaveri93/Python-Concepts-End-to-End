- `Mutable -> If we increase or decrease the Memory size `
- `immutable -> we cant increase or decrease Memory size`

**Python String concept is Mutable or Immutable Prove it**

```python
f = 'data' # save in the Memory
```

```python
# try to increase the Memory size

f = 'data' # save in the Memory

f[4] = 'j'  # want to give j to 4 th box


print(f) # calling from the Memory
```

```python
# try to delete the Memory size

f = 'data' # save in the Memory

del f[2]


print(f) # calling from the Memory
```

`Once a string save in Memory under a varibale we cant increase and we cant decrease the Memory size`

**From the Above Operations we can say that Python Strings are Immutable**

**split and join**

```python
# split

a = 'india usa uk' # save in the Memory
print(a)


b = a.split()
print(b)
```

```python
# split

a = 'india usa.uk' # save in the Memory
print(a)


b = a.split('.')
print(b)
```

```python
# join

f = ['india' , 'usa' , 'uk']

' '.join(f)
```
